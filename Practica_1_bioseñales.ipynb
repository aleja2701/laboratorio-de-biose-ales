## 1. Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
def calcularNotas(notas_alumnos):
  """
  Calcula estadísticas descriptivas de las notas de los alumnos.

  Args:
    notas_alumnos: Un diccionario donde las claves son los nombres de los alumnos y los valores son sus notas.

  Returns:
    Una serie de pandas con la nota mínima, máxima, media y desviación típica.
    Devuelve None si el diccionario está vacío o no contiene notas válidas.
  """

  df_notas = pd.DataFrame.from_dict(notas_alumnos, orient='index', columns=['nota'])

  notas = list(notas_alumnos.values())

  minima = df_notas.min().values[0]
  maxima = df_notas.max().values[0]
  media = df_notas.mean().values[0]
  media = df_notas.mean().values[0]
  desviacion_tipica = df_notas.std().values[0]

  estadisticas = pd.Series({'Mínimo': minima, 'Máximo': maxima, 'Media': media, 'Desviación típica': desviacion_tipica})
  return estadisticas

notas = {'Ana': 7, 'Luis': 8, 'Maria': 9, 'Pedro': 6, 'Sofia': 10}
estadisticas = calcularNotas(notas)
print(estadisticas)
```

### d. Calcular el IMC con la siguiente fórmula

$$
\text{IMC} = \frac{\text{Peso (kg)}}{\left(\text{Altura (m)}\right)^2}
$$


```python
# La fórmula correcta es la siguiente, pero en realidad las columnas Weight y Height están cruzadas, por lo que se van a crear nuevas columnas para mayor claridad
# df['IMC'] = df['Weight'] / ((df['Height']/100) ** 2)

df['Altura_M'] = df['Weight']/100
df['Peso_Real'] = df['Height']

df['Peso_Real'] / (df['Altura_M'] ** 2)
```

### e. Agregar el IMC como una nueva columna llamada "BMI" en el DataFrame.

```python
df['Altura_M'] = df['Weight']/100
df['Peso_Real'] = df['Height']

df['BMI'] = df['Peso_Real'] / (df['Altura_M'] ** 2)

df = df.drop(columns=['Altura_M', 'Peso_Real'])

df
```

### f. (Opcional) Categorizar a los participantes según su IMC


```python
def categorizar_imc(imc):
    if imc < 18.5:
        return 'Bajo peso'
    elif 18.5 <= imc < 25:
        return 'Normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

df['category'] = df['BMI'].apply(categorizar_imc)

df
```
